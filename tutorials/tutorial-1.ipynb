# Tutorial 1: Continuous systems 1DoF

In this very first tutorial we will introduce the basic setups for computation and visualization of the Lagrangian descriptor from simulation of the dynamics of 1DoF continious dynamical systems: autonomous and nonautonomous (time-dependent).

First, we will look at the Hamilton centre system, a well-known autonomous systems, enumarating every step.

Then, additional examples of other Hamiltonian systems are added to reinforce the logic and for you to play around with them!

## A first example: The Hamilton centre system

In this first example, we look at the centre system: a hamiltonian autnomous system described by the Hamiltonian function, identified as the energy of the system, and its vector field given by the Hamilton equations.

_Energy_

\begin{equation*}
H(x, p_x) = \frac{\omega}{2} (p_x^2 + x^2) \;, \;\; \omega > 0
\end{equation*}

_Vector field_

\begin{align*}
 \dot{x} &= \frac{\partial H}{\partial p_x} = f_1(x, p_x) = \omega p_x \\
 \dot{p}_x &= -\frac{\partial H}{\partial x} = f_2(x, p_x) = -\omega x
 \end{align*}

__FIRST__

We begin by importing the basic functions from the `ldds` module to calculate and plot the Lagrangian descriptors (LDs) of the dynamics of the system, and also the function that define its vector field.

```python
from ldds.base import compute_lagrangian_descriptor
from ldds.tools import draw_all_lds
from ldds.vector_fields import HamCenter1D
```

_NOTE_  When running this notebook, if you haven't installed `ldds` then, you might get an import error from Python. Try out the following lines to fix that, restart the notebook, and then try importing the above functions again. 

```python
import os, sys
import numpy as np
sys.path.insert(1, os.pardir)
```

__SECOND__

Define the paramaters for a 2D rectangular grid for visualization of the Lagrangian descriptor in the plane $x-p_x$, and from which initial conditions to solve the dynamics are taken too. 

```python
x_min,x_max = [-1.6, 1.6] # x axis limits
y_min,y_max = [-1.6, 1.6] # p_x axis limits
Nx, Ny = [300, 300] # Number of points per axis

grid_parameters = [(x_min, x_max, Nx), (y_min, y_max, Ny)]
```

__THIRD__

Define the paramaters for integration of the dynamics, the integration time $\tau$ and the $p$-value determining the deifnition of the Lagrangian descriptor function to use.

```python
# Integration parameters
tau = 12

# LDp, p-value
p_value = 0.5
```

_NOTE_ For further info on different versions of the Lagrangia descriptor function. Look at this online reference. [Here](https://champsproject.github.io/lagrangian_descriptors/content/chapter3.html#formulations-for-lagrangian-descriptors).

__FOURTH__

Define the vector field function describing the dynamics.

```python
vector_field = HamCenter1D
```

LDDS provides a thorough description of every vector field function, which can be displayed using `help`.

```python
help(HamCenter1D)
```

This is particularly helpful if you want to change `PARAMETERS` of the system, which can be done defining `vector_field` as below, for example:

```python
my_omega_value = 1
vector_field = lambda t, u : HamCenter1D(t, u, PARAMETERS=[my_omega_value])
```

_NOTE_ 

Here, we imported the function `HamCenter1D` that lives in the module `ldds.vector_fields`.

LDDS can also accept user-defined functions, as it will be seen later in [Tutorial 4](tutorial-4.ipynb)

__FIFTH__ 

Compute the forward and backward Lagrangian descriptors passing all the above parameters to `compute_lagrangian_descriptor`. Note that for the _backwards LD_ the integration time is passed with a negative sign, since the dynamics has to be solved "backwards in time" for this LD.

```python
LD_forward = compute_lagrangian_descriptor(grid_parameters, vector_field, tau, p_value)
LD_backward = compute_lagrangian_descriptor(grid_parameters, vector_field, -tau, p_value)
```

__SIXTH__

Visualise the Lagrangian descriptor and its gradient using the `draw_all_lds` function. 

This will return two contour plots, one of LD values and another of its gradient magnitude over each point in the 2D grid that we defined initially.

```python
figs = draw_all_lds(LD_forward, LD_backward, grid_parameters, tau, p_value)
```

__SEVENTH__ (OPTIONAL)

The graphical output provided by `draw_all_lds` can be further modified, since it returns the plot handles of the figure. So, typical modifications to Python `matplotlib` objects apply. 

Below we show how the first row of plots can be customised from the outputted subplot handles.

```python
# choose LD forwards subplots
fig, ax = figs[0]

# set new axes labels
ax[0].set_xlabel("q", fontsize=18)
ax[0].set_ylabel("p", fontsize=18)
ax[1].set_xlabel("q", fontsize=18)
ax[1].set_ylabel("p", fontsize=18)

# set new super-title
suptitle = "My LDDS output"
fig.suptitle(suptitle)

# visualise
fig
```

## More autonomous systems

### Hamilton Saddle

```python
from ldds.vector_fields import HamCenter1D, HamSaddle1D, Duffing1D
```

_Energy_

\begin{equation*}
H(x, p_x) = \frac{\lambda}{2}( p_x^2 -  x^2 ) \; , \;\; \lambda > 0
\end{equation*}

_Vector field_

\begin{align*}
 \dot{x} &= \frac{\partial H}{\partial p_x} = f_1(x, p_x) = \lambda p_x \\
 \dot{p}_x &= -\frac{\partial H}{\partial x} = f_2(x, p_x) = \lambda x
 \end{align*}

```python
# Integration parameters
tau = 12

# LDp, p-value
p_value = 0.5

# Mesh parameters
x_min,x_max = [-1.6, 1.6]
y_min,y_max = [-1.6, 1.6]
Nx, Ny = [300, 300]
```

```python
grid_parameters = [(x_min, x_max, Nx), (y_min, y_max, Ny)]
vector_field = HamSaddle1D
```

```python
LD_forward = compute_lagrangian_descriptor(grid_parameters, vector_field, tau, p_value)
LD_backward = compute_lagrangian_descriptor(grid_parameters, vector_field, -tau, p_value)
```

```python
figs = draw_all_lds(LD_forward, LD_backward, grid_parameters, tau, p_value)
```

### Duffing oscillator

_Energy_

\begin{equation*}
H(x, p_x) = \dfrac{1}{2}p_x^2 - \dfrac{1}{2}x^2 + \dfrac{1}{4}x^4
\end{equation*}

_Vector field_

\begin{align*}
 \dot{x} &= \frac{\partial H}{\partial p_x} = f_1(x, p_x) = p_x \\
 \dot{p}_x &= -\frac{\partial H}{\partial x} = f_2(x, p_x) = x - x^3
 \end{align*}

```python
# Integration parameters
tau = 12

# LDp, p-value
p_value = 0.5

# Mesh parameters
x_min,x_max = [-1.6, 1.6]
y_min,y_max = [-1, 1]
Nx, Ny = [300, 300]
```

```python
grid_parameters = [(x_min, x_max, Nx), (y_min, y_max, Ny)]
vector_field = Duffing1D
```

```python
LD_forward = compute_lagrangian_descriptor(grid_parameters, vector_field, tau, p_value)
LD_backward = compute_lagrangian_descriptor(grid_parameters, vector_field, -tau, p_value)

figs = draw_all_lds(LD_forward, LD_backward, grid_parameters, tau, p_value)
```
